```python
from llama_index.core.prompts.prompts import SimpleInputPrompt
system_prompt="""
You are a Q&A assistant. Your goal is to answer questions as accurately as possible based on the instructions and context provided.
"""

#Default format supportable by LLama2
query_wrapper_prompt=SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
from llama_index.llms.huggingface import HuggingFaceLLM
```

```python
# Create an instance of the HuggingFaceLLM class
llm1 = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",  # Specify the model name
    device_map="auto"  # Use the default device
)
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.core.service_context import ServiceContext
from langchain.embeddings import HuggingFaceEmbeddings

embed_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
```

```python
!pip install llama-index-embeddings-langchain
from llama_index.core import ServiceContext

# Create a service context object
service_context = ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm1,
    embed_model=embed_model
)
```

```python
service_context
```

```python
index=VectorStoreIndex.from_documents(documents, service_context=service_context)
```

```python
index
```

```python
query_engine=index.as_query_engine()
```

```python
response=query_engine.query("What is Termination of effect of Act")
```
